```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests,json
import datetime
from pandas import HDFStore
```

```

datelist = pd.date_range( periods=40, freq='D', end=datetime.datetime.today() ).tolist()
```

```
dfTest = None

src = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/'
for d in datelist[:-1]:
  print(src + d.strftime('%m-%d-%Y')+'.csv')
  dfData = pd.read_csv(src + d.strftime('%m-%d-%Y')+'.csv')
  dfData = dfData[(dfData.Country_Region=='US')].groupby('Province_State').sum()
  dfData = dfData[['Confirmed', 'Deaths','Active']]
  dfData['date'] = datetime.datetime(d.year, d.month, d.day)

  
  if dfTest is None:
    dfTest = dfData
  else:
    dfTest = pd.concat([dfTest,dfData])
```
